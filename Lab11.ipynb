Transposed convolution is used for upsampling (created output feature map has a larger spatial dimension). Let's start by demonstrating how it works on an example.

IMG_6730 (1).JPG

As we saw, for each element in the input grid we multiply it by the kernel and place the result in the output grid, next we sum each resulting grid.\
Stride corresponds to the length of steps we take while placing each kernel multiplication result.\
Padding is used to remove the border of the output grid.\
Output size can be computed using:
$$O=S\cdot (I−1)+K−2P,$$ where
I = input size,
K = kernel size,
S = stride,
P = padding,
O = output size.
Using last example we get $$(2,2)\cdot ((2,2)-(1,1))+(1,2)-2\cdot 0=(2,2)+(1,2)=(3,4).$$
